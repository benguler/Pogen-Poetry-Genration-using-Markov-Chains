### opens a csv file containing one field such as the 'content' of a poem
### read the csv file line by line 
### and write each line plus a lable to a new csv file
### e.g. Love is abudant like oxygen,modern_love

```python
import csv

filename = 'r_nature_proc.csv'
TAG = 'renaissance_nature'
new_filename = 'renaissance_nature7_proc.csv'

with open('CSVs/'+filename, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file, delimiter='\n')
    
    #next(csv_reader)
    
    with open ('CSVs/processed/'+new_filename, 'w') as new_file:
        fieldnames = ['poem_line', TAG]
        csv_writer = csv.writer(new_file)
        
        csv_writer.writerow(fieldnames)
        
        for line in csv_reader:
            # write the the contents of the line['content'], and add a tag separated by a comma for each line 
            #print([line[TAG]] + [TAG])            
            csv_writer.writerow([line[TAG]] + [TAG])

            
        print("wrote to csv file: " + new_filename)
```

### Opens a processed csv file and uses Naive Bayes Classifier to train a model

```python
from textblob.classifiers import NaiveBayesClassifier
import csv


with open('CSVs/processed/renaissance_nature_proc.csv', 'r') as csv_file2:
    csv_reader = csv.reader(csv_file2, delimiter =',')
    cl = NaiveBayesClassifier(csv_file2, format="csv")
#     for line in csv_reader:
#         cl.classify()
    #print('With file: ' + new_filename)
    print(cl)
```

```python
proba = cl.prob_classify("A church bell sounded mournfully far away")

print(round(proba.prob('And make me travel forth without my cloak'), 2))
```

### Concatenate all csv files

```python
import csv
file_list = ['CSVs/processed/modern_love_proc.csv',
             'CSVs/processed/modern_mythology_folklore_proc.csv',
             'CSVs/processed/modern_nature_proc.csv', 
             'CSVs/processed/renaissance_love_proc.csv',
             'CSVs/processed/renaissance_mythology_folklore_proc.csv',
             'CSVs/processed/renaissance_nature_proc.csv']


for i in range(len(file_list)):
    with open(file_list[i], 'r') as the_file:
        csv_reader = csv.reader(the_file)
        next(csv_reader)
        with open('CSVs/processed/all_proc.csv', 'a+') as big_file:
            #fieldname = ['poem_line', 'poem_genre']
            csv_writer = csv.writer(big_file)
            
            for line in csv_reader:
                csv_writer.writerow(line)
                #print(line)

        the_file.close()
        
        
```

### Tokenization of poems words into a nested list

```python
import csv
from textblob import TextBlob
#from tokenizers import WordTokenizer
list_1 = []
matrix = []
with open ('CSVs/processed/all_proc.csv', 'r') as csvfile:
    csv_reader = csv.DictReader(csvfile)
    data_lines = list(csv_reader)

    for line in data_lines:
        poem_sentence = TextBlob(line['poem_line'])
        mystr = poem_sentence.words
        matrix.append(mystr)
        
print(matrix)
```
