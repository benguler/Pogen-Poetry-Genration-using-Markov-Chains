opens a csv file containing one field such as the 'content' of a poem
read the csv file line by line 
and write each line plus a lable to a new csv file
e.g. Love is abudant like oxygen,modern_love

```python
import csv
with open('CSVs/test.csv', 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file, delimiter='\n')
    
    #next(csv_reader)
    
    with open ('CSVs/processed/test_proc.csv', 'w') as new_file:
        fieldnames = ['poem_line', 'test']
        csv_writer = csv.writer(new_file)
        
        csv_writer.writerow(fieldnames)
        
        for line in csv_reader:
            # write the the contents of the line['content'], and add a tag separated by a comma for each line 
            csv_writer.writerow([line['test']] + ['test'])
```

Opens a processed csv file and uses Naive Bayes Classifier to train a model

```python
from textblob.classifiers import NaiveBayesClassifier
import csv

with open('CSVs/processed/modern_mythology_folklore_proc.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter =',')
    cl = NaiveBayesClassifier(csv_file, format="csv")
    for line in csv_reader:
        cl.classify()
    print(cl)
```

```python
proba = cl.prob_classify("A church bell sounded mournfully far away")

print(round(proba.prob('And make me travel forth without my cloak'), 2))
```
