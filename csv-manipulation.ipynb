This file opens the main csv file (all.csv) and write to a new csv file according to the given fields

filename - the name of the fiile

AGE - Modern or Renaissance

Type - Nature, Love, or Mythology & Folklore

The code reads a poem by poem ('content') and writes to a new csv along with a TAG


e.g. 

This is a poem,
Roses are red, violets are blue
programming in python all the way to the moon.,modern_love



```python
from textblob import TextBlob
import csv
data = open('all.csv', encoding="utf-8")
csv_data = csv.reader(data)
data_lines = list(csv_data)
fields = ['content', 'tag']
filename = 'test.csv'
AGE = 'Renaissance'
TYPE = 'Mythology & Folklore'
TAG_NAME = 'renaissance_mythology_folklore'
```

```python
with open(filename, 'w') as csvfile:
    #creates a csv writer object
    csv_writer = csv.writer(csvfile)
    
    #writing the fields
    csv_writer.writerow(fields)
    
    #writing rows
    for line in data_lines[1::2]:
        if(line[3] == AGE and line[4] == TYPE):
            print(line[1] + ','+TAG_NAME+'\n\n')
            #csvfile.write(line[1] + ','+TAG_NAME+'\n\n')

```
