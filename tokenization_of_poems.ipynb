```python
from textblob import TextBlob
import csv
data = open('all.csv', encoding="utf-8")
csv_data = csv.reader(data)
data_lines = list(csv_data)
```

Populate set with words

```python
word_set = set()


for line in data_lines:
    poem_line = TextBlob(line[1])
    for s in poem_line.words:
        if (len(s) > 1):
            word_set.add(s)
        
        
    
    
```

Attempt to populate list with sentences

```python
sentence_list = []

for line in data_lines:
    #print({}, line[1])
    poem = TextBlob(line[1])
    for s in poem:
        poem_line += s
        if (s == '\n'):
            sentence_list.append(poem_line)
            poem_line = ""
        
        #print(s)
    #sentence_list.append(poem_line.sentences)
    
```
